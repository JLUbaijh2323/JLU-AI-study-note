```python
import random
random.seed(2021)

xi = [random.randint(0,100) for i in range(10)]
xj = [random.randint(0,100) for j in range(10)]

for p in range(1, 101):
    # 计算 |xi - xj|^p 的和
    resu = sum(abs(xi[u]-xj[u])**p for u in range(len(xi)))
    # p 次方根
    result = resu ** (1/p)
    print(f"p={p}, 距离={result}")
```
