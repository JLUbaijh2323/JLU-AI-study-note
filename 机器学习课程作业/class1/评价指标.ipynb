```python
import random
random.seed(2022)
# 生成概率和标签
prob = [random.random() for i in range(100)]
label = [random.randint(0,1) for i in range(100)]
# 初始化混淆矩阵
TP, FP, FN, TN = 0, 0, 0, 0
# 阈值固定 0.5
for x, y in zip(prob, label):
    if x > 0.5 and y == 1:
        TP += 1
    elif x > 0.5 and y == 0:
        FP += 1
    elif x <= 0.5 and y == 1:
        FN += 1
    else:  # x <= 0.5 and y == 0
        TN += 1
# 总数
N = TP + FP + FN + TN
# 各类指标
ERR = (FP + FN) / N
ACC = (TP + TN) / N
Pre = TP / (TP + FP) if (TP+FP) > 0 else 0
Recall = TP / (TP + FN) if (TP+FN) > 0 else 0
F1 = 2 * Pre * Recall / (Pre + Recall) if (Pre+Recall) > 0 else 0

print(f"ERR = {ERR:.4f}")
print(f"ACC = {ACC:.4f}")
print(f"Precision = {Pre:.4f}")
print(f"Recall = {Recall:.4f}")
print(f"F1 = {F1:.4f}")
```

```python
import random
import numpy as np
from sklearn.metrics import roc_auc_score

random.seed(2022)

# 随机生成预测概率和标签
prob = [random.random() for i in range(100)]
label = [random.randint(0,1) for i in range(100)]

# 手算 AUC 
X, Y, AUC = 0, 0, 0

for R in np.linspace(1,0,1001):  # 遍历阈值
    TP = FP = FN = TN = 0  # 每个阈值重置混淆矩阵
    for x,y in zip(prob,label):
        if x > R and y == 1:
            TP += 1
        elif x > R and y == 0:
            FP += 1
        elif x <= R and y == 1:
            FN += 1
        else:  # x <= R and y == 0
            TN += 1

    # 计算 TPR / FPR
    FPR = FP/(FP+TN) if (FP+TN)>0 else 0
    TPR = TP/(TP+FN) if (TP+FN)>0 else 0

    # 梯形面积累加
    AUC += (FPR - X) * (TPR + Y) / 2
    X, Y = FPR, TPR

print(f"手算 AUC ={AUC:.4f}",)

# sklearn AUC
auc_sklearn = roc_auc_score(label, prob)
print(f"sklearn AUC ={auc_sklearn:.4f}" )

# 对比
print("差值 =", abs(AUC - auc_sklearn))
```
