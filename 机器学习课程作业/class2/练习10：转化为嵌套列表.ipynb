```python
import pickle
# 数据转换
data = []
with open('ML100k.data','r') as file:
    for line in file:
        u, m, r, t = map(int, line.strip().split())
        data.append([u, m, r, t])
print("前10条：", data[:10])

with open('ML100k.pkl', 'wb') as file:
    pickle.dump(data, file)

with open('ML100k.pkl','rb') as file:
    data = pickle.load(file)
```
