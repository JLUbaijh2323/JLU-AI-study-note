```python
import random

times = 1000
result = {'Heart':0,'Queen':0,'Heart or Queen':0}
poker = []

for i in range(4):
    for j in range(13):
        poker.append([i,j])
for i in range(times):
    k = random.choice(poker)
    if k[0] == 0:
        result['Heart']+=1
for i in range(times):
    k = random.choice(poker)
    if k[1] == 12:
        result['Queen']+=1
for i in range(times):
    k = random.choice(poker)
    if k[0] == 11 or k[1]==12:
        result['Heart or Queen']+=1
print("Times of Heart = %d \
        Time of Queen = %d \
        Time of Heart or Queen = %d \
        Probability of Heart = %f \
        Probability of Queen = %f \
        Probability of Heart or Queen = %f \
        "%(result['Heart'],result['Queen'],result['Heart or Queen'],result['Heart']/times,result['Queen']/times,result['Heart or Queen']/times))
```

```python
import random

# 1. 构建扑克牌数据结构
def create_deck():
    styles = ['red_heart', 'black_spade', 'red_diamond', 'black_club']
    keys = list(range(1, 14))
    deck = []
    for style in styles:
        for key in keys:
            deck.append({'style': style, 'key': key})
    return deck

# 2. 设计核心函数：模拟一次摸牌实验
def simulate_draws(deck, num_draws):
    # 初始化统计结果
    results = {
        'red_heart_count': 0,
        'Q_count': 0,
        'red_heart_or_Q_count': 0
    }
    
    for _ in range(num_draws):
        # 从牌堆中随机抽取一张牌
        card = random.choice(deck)
        
        # 统计红桃出现的次数
        if card['style'] == 'red_heart':
            results['red_heart_count'] += 1
            
        # 统计Q出现的次数 (Q 对应点数 12)
        if card['key'] == 12:
            results['Q_count'] += 1
            
        # 统计红桃或者Q出现的次数
        if card['style'] == 'red_heart' or card['key'] == 12:
            results['red_heart_or_Q_count'] += 1
            
    return results

# 3. 主程序流程
if __name__ == "__main__":
    deck = create_deck()
    times_to_run = [1000, 10000, 100000]
    
    for times in times_to_run:
        print(f"--- 模拟 {times} 次实验 ---")
        simulation_results = simulate_draws(deck, times)
        # 打印结果
        print(f"红桃出现次数: {simulation_results['red_heart_count']}")
        print(f"Q 出现次数: {simulation_results['Q_count']}")
        print(f"红桃或Q出现次数: {simulation_results['red_heart_or_Q_count']}")
        print("---")
```
